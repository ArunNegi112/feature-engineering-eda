```python
import numpy as np
import seaborn as sns
```

5 number summary: Minimum, Q1(25%ile), Median, Q3(75%ile), Maximum  
IQR (interquartile range) = Q3 - Q1  
Outliers: values below (Q1 - 1.5 * IQR) or above (Q3 + 1.5 * IQR)  

```python
lst = [10,12,25,20,1,-10,20,49,50,60,34,53,35,44,20]
print(sorted(lst))

min,Q1,med,Q3,max = np.quantile(lst,[0,0.25,0.50,0.75,1.0])
tuple(map(int,[min,Q1,med,Q3,max]))
```

```python
# Getting outliers
IQR = Q3-Q1

lower_fence = Q1 - 1.5*IQR
higher_fence = Q3 + 1.5*IQR

float(lower_fence),float(higher_fence)
# anything lower than lower fence is outlier and higher than higher fence is outlier
```

```python
sns.boxplot(lst)
```

Boxplot can directly show outliers and 5 number summary
- lowest = minimum
- lower edge of box = Q1
- middle line of box = median
- upper edge of box = Q3
- highest = maximum
