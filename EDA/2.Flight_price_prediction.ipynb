### Applying feature engineering

```python
# Creating a Dataframe for encoded data
encoded_Df = pd.DataFrame()
```

```python
# Encoding 'Airline' feature
data['Airline'].value_counts()
```

```python
from sklearn.preprocessing import OneHotEncoder

encode_airline = OneHotEncoder()
airline_encoded = pd.DataFrame(encode_airline.fit_transform(pd.DataFrame(data['Airline'])).toarray(),columns=encode_airline.get_feature_names_out())

encoded_Df = pd.concat([encoded_Df,airline_encoded],axis=1)
encoded_Df
```

```python
data['Date_of_Journey'].str.split("/")
```

```python
# converting 'Date of journey' as integer, for that
# we need to separate day month and year


# Method 1
encoded_Df['Date'] = data['Date_of_Journey'].str.split("/").str[0]
encoded_Df['Month'] = data['Date_of_Journey'].str.split("/").str[1]
encoded_Df['Year'] = data['Date_of_Journey'].str.split("/").str[2]

encoded_Df.head()
```

```python
encoded_Df[['Date','Month','Year']].info()

# But still we see that it's as 'object' type, gotta convert it into 'int'
```

```python
encoded_Df[['Date','Month','Year']]=encoded_Df[['Date','Month','Year']].map(int)
encoded_Df[['Date','Month','Year']].info()
```

```python
# Method 2 (Better approach)

# convert the dtype into datetime
data['Date_of_Journey'] = pd.to_datetime(data['Date_of_Journey'],format="%d/%m/%Y")

''' Then add it in 'encoded_df' '''
# encoded_Df['day'] = data['Date_of_Journey'].dt.day
# encoded_Df['month'] = data['Date_of_Journey'].dt.month
# encoded_Df['year'] = data['Date_of_Journey'].dt.year

# encoded_Df
```

```python
# Encoding 'Source' and 'Destination' Together

encode_source = OneHotEncoder()
encoded_source_df = pd.DataFrame(encode_source.fit_transform(pd.DataFrame(data[['Source','Destination']])).toarray(),columns=encode_source.get_feature_names_out())
encoded_Df=pd.concat([encoded_Df,encoded_source_df],axis=1)
encoded_Df
```

```python
# converting 'Dep_time' as integer, for that
# we need to separate hour and min

data['Dep_Time'] = pd.to_datetime(data['Dep_Time'],format="%H:%M")

encoded_Df['Dep_Hour'] = data['Dep_Time'].dt.hour
encoded_Df['Dep_min'] = data['Dep_Time'].dt.minute

encoded_Df.head()
```

```python
# converting 'Arrival_time' as integer, for that
# we need to separate hour and min and remove that date there

data['Arrival_Time'] = data['Arrival_Time'].map(lambda i:i.split(" ")[0])     # Removed that date
data['Arrival_Time'] = pd.to_datetime(data['Arrival_Time'],format="%H:%M")    # coverted into datetime format

encoded_Df['Arrival_hour'] = data['Arrival_Time'].dt.hour
encoded_Df['Arrival_min'] = data['Arrival_Time'].dt.minute

encoded_Df[['Arrival_hour','Arrival_min']].head()
```

```python
encoded_Df.head(2)
```

```python
## To convert 'Duration' in integer, 
# We converting it into minutes

hour_series =  data['Duration'].map(lambda i:i.split(" ")).str[0].str.removesuffix("h")
min_series = data['Duration'].map(lambda i:i.split(" ")).str[1].str.removesuffix("m").fillna(0)

hour_series.unique()  # Notice that, that '5m' is causing problems
```

```python
print(hour_series[hour_series=='5m'])
```

```python
hour_series.drop(hour_series[hour_series=='5m'].index,inplace=True)
```

```python
hour_series.unique()  # No '5m' anymore, It can also be considered as outlier too (thats why we dropped it)
```

```python
# Converting hour_class and min_class as integer
hour_series = hour_series.map(int)
min_series = min_series.map(int)

print(hour_series.dtype,min_series.dtype)
```

```python
encoded_Df['Duration'] = 60*hour_series + min_series
encoded_Df.head(2)
```

```python
## Ordinally encoding Total_Stops (because taking 3Stops cost more than 2Stops )

data['Total_Stops'] = data['Total_Stops'].map(lambda i:str(i).split(" ")[0])
print(data['Total_Stops'].head(3))    # To remove those 'non-stop'

data['Total_Stops'].unique()
```

```python
# To replace those "nan" values we use mode of Total_stops
mode = data['Total_Stops'].mode(dropna=True)
encoded_Df['Total_steps'] = data['Total_Stops'].map({'non-stop':0,'1':1,'2':2,'3':3,'4':4,'nan':mode[0]})
```
